```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
data = pd.read_csv('/content/Uber-Jan-Feb-FOIL.csv')

# Display basic info about the dataset
print(data.info())

# Data Preprocessing
# Convert 'date' to datetime object
data['date'] = pd.to_datetime(data['date'])
# Extracting useful information from 'date'
data['Hour'] = data['date'].dt.hour
data['Day'] = data['date'].dt.day
data['DayOfWeek'] = data['date'].dt.dayofweek
data['Month'] = data['date'].dt.month

# Exploratory Data Analysis
# Plotting the number of trips per hour
plt.figure(figsize=(10, 6))
sns.countplot(x='Hour', data=data)  # Changed to use countplot with x and data parameters
plt.title('Trips per Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Trips')
plt.show()

# Plotting the number of trips per day of the week
plt.figure(figsize=(10, 6))
sns.countplot(x='DayOfWeek', data=data)  # Changed to use countplot with x and data parameters
plt.title('Trips per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Trips')
plt.show()

# Feature Engineering
# Create dummy variables for categorical features
data = pd.get_dummies(data, columns=['dispatching_base_number'], drop_first=True)

# Define features and target variable
# Exclude 'Lat' and 'Lon' as they are not present in the data
X = data[['Hour', 'Day', 'DayOfWeek', 'Month']]
y = data['trips']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Model Building
# Train a Random Forest Regressor
rfr = RandomForestRegressor(random_state=42)
rfr.fit(X_train, y_train)

# Predict on the test set
y_pred = rfr.predict(X_test)

# Model Evaluation
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R^2 Score:", r2_score(y_test, y_pred))

# Visualization of Predictions
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.3)
plt.xlabel('Actual Trips')
plt.ylabel('Predicted Trips')
plt.title('Actual vs Predicted Trips')
plt.show()
```

```python
# 1. Import Libraries
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import xgboost as xgb

from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.model_selection import TimeSeriesSplit, train_test_split
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

# 2. Helper Functions
def PlotDecomposition(result):
    plt.figure(figsize=(22,18))
    plt.subplot(4,1,1)
    plt.plot(result.observed, label='Observed', lw=1)
    plt.legend(loc='upper left')
    plt.subplot(4,1,2)
    plt.plot(result.trend, label='Trend', lw=1)
    plt.legend(loc='upper left')
    plt.subplot(4,1,3)
    plt.plot(result.seasonal, label='Seasonality', lw=1)
    plt.legend(loc='upper left')
    plt.subplot(4,1,4)
    plt.plot(result.resid, label='Residuals', lw=1)
    plt.legend(loc='upper left')
    plt.tight_layout()
    plt.show()

def create_lagged_features(data, window_size):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size])
    return np.array(X), np.array(y)

def PlotPredictions(plots, title):
    plt.figure(figsize=(18, 8))
    for plot in plots:
        plt.plot(plot[0], plot[1], label=plot[2], linestyle=plot[3], color=plot[4], lw=1)
    plt.xlabel('Date')
    plt.ylabel("Trips")
    plt.title(title)
    plt.legend()
    plt.xticks(rotation=30, ha='right')
    plt.tight_layout()
    plt.show()

# 3. Load Dataset
df = pd.read_csv("Uber-Jan-Feb-FOIL.csv")
df['date'] = pd.to_datetime(df['date'])
daily_trips = df.groupby('date')['trips'].sum().reset_index()
daily_trips = daily_trips.sort_values(by='date')
daily_trips.set_index('date', inplace=True)

# 4. Time Series Decomposition
result = seasonal_decompose(daily_trips['trips'], model='additive', period=7)
PlotDecomposition(result)

# 5. Train-Test Split
cutoff_date = '2015-02-01'
train_data = daily_trips.loc[:cutoff_date]
test_data = daily_trips.loc[cutoff_date:]

# 6. Create Lagged Features
window_size = 7
X_train, y_train = create_lagged_features(train_data['trips'].values, window_size)
test_series = np.concatenate([train_data['trips'].values[-window_size:], test_data['trips'].values])
X_test, y_test = create_lagged_features(test_series, window_size)

# 7. XGBoost Model
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
xgb_model.fit(X_train, y_train)
xgb_predictions = xgb_model.predict(X_test)
xgb_mape = mean_absolute_percentage_error(y_test, xgb_predictions)

# 8. Random Forest Model
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
rf_mape = mean_absolute_percentage_error(y_test, rf_predictions)

# 9. Gradient Boosting Model
gbr_model = GradientBoostingRegressor(random_state=42)
gbr_model.fit(X_train, y_train)
gbr_predictions = gbr_model.predict(X_test)
gbr_mape = mean_absolute_percentage_error(y_test, gbr_predictions)

# 10. Ensemble Model
weights = np.array([1/xgb_mape, 1/rf_mape, 1/gbr_mape])
weights /= weights.sum()
ensemble_predictions = (weights[0] * xgb_predictions +
                        weights[1] * rf_predictions +
                        weights[2] * gbr_predictions)
ensemble_mape = mean_absolute_percentage_error(y_test, ensemble_predictions)

# 11. Plot Predictions
PlotPredictions([
    (test_data.index[:len(y_test)], y_test, 'Actual', '-', 'black'),
    (test_data.index[:len(xgb_predictions)], xgb_predictions, 'XGBoost', '--', 'red'),
    (test_data.index[:len(rf_predictions)], rf_predictions, 'Random Forest', '--', 'green'),
    (test_data.index[:len(gbr_predictions)], gbr_predictions, 'GBRT', '--', 'orange'),
    (test_data.index[:len(ensemble_predictions)], ensemble_predictions, 'Ensemble', '--', 'purple'),
], 'Model Predictions vs Actual')

# 12. Print MAPE Scores
print("XGBoost MAPE: {:.2f}%".format(xgb_mape * 100))
print("Random Forest MAPE: {:.2f}%".format(rf_mape * 100))
print("GBRT MAPE: {:.2f}%".format(gbr_mape * 100))
print("Ensemble MAPE: {:.2f}%".format(ensemble_mape * 100))
```

```python
import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import xgboost as xgb

from statsmodels.tsa.seasonal import seasonal_decompose
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.model_selection import TimeSeriesSplit
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

# Plot seasonal decomposition results
def PlotDecomposition(result):
    plt.figure(figsize=(22,18))
    plt.subplot(4,1,1)
    plt.plot(result.observed, label='Observed', lw=1)
    plt.legend(loc='upper left')
    plt.subplot(4,1,2)
    plt.plot(result.trend, label='Trend', lw=1)
    plt.legend(loc='upper left')
    plt.subplot(4,1,3)
    plt.plot(result.seasonal, label='Seasonality', lw=1)
    plt.legend(loc='upper left')
    plt.subplot(4,1,4)
    plt.plot(result.resid, label='Residuals', lw=1)
    plt.legend(loc='upper left')
    plt.tight_layout()
    plt.show()

# Create lagged features for time series modeling
def create_lagged_features(data, window_size):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i+window_size])
        y.append(data[i+window_size])
    return np.array(X), np.array(y)

# Plot actual vs predicted trips
def PlotPredictions(plots, title):
    plt.figure(figsize=(18, 8))
    for plot in plots:
        plt.plot(plot[0], plot[1], label=plot[2], linestyle=plot[3], color=plot[4], lw=1)
    plt.xlabel('Date')
    plt.ylabel("Trips")
    plt.title(title)
    plt.legend()
    plt.xticks(rotation=30, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# Load Uber dataset (must be in the same directory)
df = pd.read_csv("Uber-Jan-Feb-FOIL.csv")

# Convert to datetime
df['date'] = pd.to_datetime(df['date'])

# Aggregate total trips per day
daily_trips = df.groupby('date')['trips'].sum().reset_index()
daily_trips = daily_trips.sort_values(by='date')
daily_trips.set_index('date', inplace=True)
```

```python
# Decompose time series to observe trend, seasonality, residuals
result = seasonal_decompose(daily_trips['trips'], model='additive', period=7)
PlotDecomposition(result)
```

```python
# Use Feb 1, 2015 as cutoff date
cutoff_date = '2015-02-01'
train_data = daily_trips.loc[:cutoff_date]
test_data = daily_trips.loc[cutoff_date:]
```

```python
# Set window size (1 week)
window_size = 7

# Prepare training data
X_train, y_train = create_lagged_features(train_data['trips'].values, window_size)

# Concatenate last week of train with test for proper windowing
test_series = np.concatenate([train_data['trips'].values[-window_size:], test_data['trips'].values])
X_test, y_test = create_lagged_features(test_series, window_size)
```

```python
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)
xgb_model.fit(X_train, y_train)
xgb_predictions = xgb_model.predict(X_test)
xgb_mape = mean_absolute_percentage_error(y_test, xgb_predictions)
```

```python
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
rf_mape = mean_absolute_percentage_error(y_test, rf_predictions)
```

```python
gbr_model = GradientBoostingRegressor(random_state=42)
gbr_model.fit(X_train, y_train)
gbr_predictions = gbr_model.predict(X_test)
gbr_mape = mean_absolute_percentage_error(y_test, gbr_predictions)
weights = np.array([1/xgb_mape, 1/rf_mape, 1/gbr_mape])
weights /= weights.sum()

ensemble_predictions = (weights[0] * xgb_predictions +
                        weights[1] * rf_predictions +
                        weights[2] * gbr_predictions)

ensemble_mape = mean_absolute_percentage_error(y_test, ensemble_predictions)

```

```python
PlotPredictions([
    (test_data.index[:len(y_test)], y_test, 'Actual', '-', 'black'),
    (test_data.index[:len(xgb_predictions)], xgb_predictions, 'XGBoost', '--', 'red'),
    (test_data.index[:len(rf_predictions)], rf_predictions, 'Random Forest', '--', 'green'),
    (test_data.index[:len(gbr_predictions)], gbr_predictions, 'GBRT', '--', 'orange'),
    (test_data.index[:len(ensemble_predictions)], ensemble_predictions, 'Ensemble', '--', 'purple'),
], 'Model Predictions vs Actual')
```
