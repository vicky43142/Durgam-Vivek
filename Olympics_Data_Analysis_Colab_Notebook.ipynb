```python

q6_data = data[['Year', 'Country', 'Medal']]
q6_data = q6_data.groupby(['Year', 'Country', 'Medal'])['Country'].count().reset_index(name='Count')
q6_data['Medal'] = pd.Categorical(q6_data['Medal'], categories=['Gold', 'Silver', 'Bronze'], ordered=True)
q6_data = q6_data.sort_values(ascending=[True, True, True], by=['Year', 'Country', 'Medal'])
q6_data = q6_data.pivot(index=['Year', 'Country'], columns=['Medal'], values=['Count']).reset_index().fillna(0)
q6_data.columns = q6_data.columns.droplevel(0)
q6_data.columns = ['Year', 'Country', 'Gold', 'Silver', 'Bronze']
q6_data['Sum'] = q6_data['Gold'] + q6_data['Silver'] + q6_data['Bronze']
q6_data.head()
```

```python

q7_data = data.groupby(['Sport', 'Country'])['Country'].count().reset_index(name='Count').sort_values(by=['Sport', 'Count'], ascending=[True, False])
q7_data.groupby('Sport').head(1)
```

```python

temp = data[['Athlete','Sport']].drop_duplicates()
temp = temp.groupby(['Athlete'])
athletes_changed_sport = [(k, v['Sport'].tolist()) for k,v in temp if len(v['Sport'].tolist()) > 1]
athletes_changed_sport[:5]
```

```python

q9_data = q6_data[['Year', 'Country', 'Sum']].groupby(['Year']).apply(lambda x: x.nlargest(5, 'Sum')).reset_index(drop=True)
top_countries = q9_data['Country'].drop_duplicates()
temp_q9 = q6_data[q6_data['Country'].isin(top_countries)][['Year', 'Country', 'Sum']]
temp_q9 = temp_q9.pivot(index='Year', columns='Country', values='Sum').fillna(0).reset_index()
temp_q9['Germany'] = temp_q9.get('Germany', 0) + temp_q9.get('East Germany', 0) + temp_q9.get('West Germany', 0)
temp_q9['Russia'] = temp_q9.get('Russia', 0) + temp_q9.get('Soviet Union', 0) + temp_q9.get('Unified team', 0)
drop_cols = ['East Germany', 'West Germany', 'Soviet Union', 'Unified team']
temp_q9.drop(columns=[c for c in drop_cols if c in temp_q9.columns], inplace=True)
temp_q9 = temp_q9.set_index('Year')

plt.figure(figsize=(15, 8))
sns.lineplot(data=temp_q9)
plt.title('Top Countries Medal Count Over Years')
plt.ylabel('Total Medals (Individuals)')
plt.xlabel('Year')
plt.grid(True)
plt.show()
```
